### Sampling

```python
from sklearn.model_selection import train_test_split
```

```python
x=cr.drop("Class",axis=1)
y=cr["Class"]
```

```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.metrics import *
```

Normalize the feature matrix

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

Handling class imbalance using SMOTE (oversampling minority class)

```python
smt = SMOTE(random_state=100)
x_resampled, y_resampled = smt.fit_resample(x_scaled,y)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x_resampled,y_resampled,test_size=.2,random_state=100)
```

### Model Building

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(train_x,train_y)
```

```python
pred_log=logreg.predict(test_x)
pred_log
```

```python
tab=confusion_matrix(test_y,pred_log)
tab
```

```python
print(classification_report(test_y,pred_log))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(train_x,train_y)
```

```python
pred_dt=dt.predict(test_x)
pred_dt
```

```python
tab1=confusion_matrix(test_y,pred_dt)
tab1
```

```python
print(classification_report(test_y,pred_dt))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(train_x,train_y)
```

```python
pred_rf=rf.predict(test_x)
pred_rf
```

```python
tab2=confusion_matrix(test_y,pred_rf)
tab2
```

```python
print(classification_report(test_y,pred_rf))
```
